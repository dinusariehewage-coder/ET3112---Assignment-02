```python


import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files


uploaded = files.upload()


img = cv.imread('/content/1c.png', 0)


edges = cv.Canny(img, 550, 690)


indices = np.where(edges != 0)
x = indices[1]
y = indices[0]


m_ls, c_ls = np.polyfit(x, y, 1)

print("Slope (LS):", m_ls)
print("Intercept (LS):", c_ls)



theta_ls = np.degrees(np.arctan(m_ls))

print("Estimated Angle (Least Squares) =", theta_ls, "degrees")


x_line = np.linspace(min(x), max(x), 100)
y_line = m_ls * x_line + c_ls

plt.figure(figsize=(6,6))
plt.scatter(x, y, s=1, label="Edge Points")
plt.plot(x_line, y_line, color='red', linewidth=2, label="Least Squares Fit")

plt.title("Least Squares Line Fit")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```
